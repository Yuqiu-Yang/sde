```python
from sde.sde_class import sde_class
import numpy as np 
from matplotlib import pyplot as plt
```

```python
sde = sde_class(T=1, N=1000, M=2)
```

```python
plt.style.use('ggplot')
for path in range(2):
    plt.plot(sde.time, sde.W[path,:],
             label="Path "+str(path))
plt.legend()
plt.xlabel("Time")
plt.ylabel("W(t)")
plt.tight_layout()
```

```python
sde = sde_class(T=1, N=1000, M=1000)
def u(t, w):
    return np.exp(t + 0.5 * w)
```

```python
transformed_W = sde.transform_W(fun=u)
```

```python
for path in range(5):
    plt.plot(sde.time, transformed_W[path,:],
             label="Path "+str(path),
             linewidth=0.5, alpha=0.7)

plt.plot(sde.time, np.mean(transformed_W, axis=0),
         linewidth=3, c='magenta',
         label="Average")
plt.plot(sde.time, np.exp(9*sde.time/8),
         linewidth=3, c='cyan',
         label="Expectation")

plt.legend()
plt.xlabel("Time")
plt.ylabel("W(t)")
plt.tight_layout()
```
